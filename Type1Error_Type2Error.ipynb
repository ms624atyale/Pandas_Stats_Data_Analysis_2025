# **Rejecting a (True) _H0_**
<font color = 'pink'> _**"There is no difference in cognitive function between patients who take the new Alzheimer’s drug and patients who take a placebo."**_

# **Type II Error(제2종 오류), β, which is also known as False Negative (위음성)😱**

#↖️ **The Sick Man with a Negative result**

Imagine you are actually infected with the virus, but you take the test too early or the swab didn't collect enough material.

The Reality: You have the COVID-19 virus in your body.

The Test Result: The kit shows only one line (C). It says you are Negative (-).

The Error: This is a False Negative (위음성). The test told you that you were healthy, but it was wrong.

The Risk: Because you believe the test, you go to school or meet friends, accidentally spreading the virus because the "alarm" failed to go off.

>

#**Accepting a (Flase) _H0_**
<font color = 'pink'> _**"There is no difference in cognitive function between patients who take the new Alzheimer’s drug and patients who take a placebo."**_

- You fail to reject(=accept) the Null Hypothesis.
- **The Consequence**: A potentially life-saving medicine is abandoned or "dies" in the lab. The pharmaceutical company loses its investment, and patients lose a treatment option.

-How to avid Type II error?
  - Increase sample size.
  - Reduce subject variability.

In stats, we use 1(certainty 필연)-β (accepting a false _H0_) to indicate the power of the test(검정력).

###**Why do we use Power?**

- <font color = 'blue'> We use Power to ensure our research is worth doing. If an experiment has low power (e.g., 20%), it means that even if your theory is 100% correct, you only have a 20% chance of actually "proving" it with your data. The other 80% of the time, you will get a non-significant result (_p > .05_) and waste your time.
- <font color = 'blue'> To avoid "The Miss": High power gives you confidence that if there is a difference, you will actually see it.

 **When do we use Power?**
i) **Before the experiment (A Priori Power Analysis)**: This is the most "correct" way to use it. You decide on your desired power (usually 0.80 or 80%) and use it to calculate how many subjects (_N_) you need. Example: "I want an 80% chance to detect a medium-sized improvement in my students. Based on a power analysis, I need at least 30 students, not just 9."

ii) **After the experiment (Post-hoc Power Analysis)**: If you get a non-significant result (like your experimentpre effect where p = .72), you might calculate the power to see if the failure was because there was no effect, or because your "net" (sample size) was too small to get the effect.

##**The Four Pillars of Power Analysis**

To find the required sample size (_N_), we must plug the following into a power calculation:

- **Significance Level (_α_)**: Usually set at 0.05. This is your tolerance for a "False Positive."
- **Desired Power (_1 - β_)**: Usually set at 0.80. This is your required sensitivity to avoid a "Miss"실책.
- **Effect Size (_d_ or f^2_)**: This is the most critical part. It represents how "big" or "noticeable" the difference is between your groups (e.g., the gap between Pre-session and Post-session scores).
- **Sample Size (_N_)**: This is what we are solving for.

##**How we estimate "Effect Size"?**
You might wonder, "How do I know the effect size before I even do the experiment?" <font color = 'red'>

**Researchers typically get this number in two ways:**

- Previous Literature: You look at similar studies in the fields of your study. If they found a "medium" effect ($d = 0.5$), you use that.
- Cohen’s Standards: Jacob Cohen, a famous statistician, provided "rule of thumb" values.
    - Small Effect ($d = 0.2$) **<font color = 'purple'> [e.g., gender difference in Math score]</font>: Requires a very large _N_ (e.g., _N > 300_).
    - Medium Effect ($d = 0.5$): Requires a moderate _N_ (e.g., _N is approximately 64_ for a t-test).
    - Large Effect ($d = 0.8$)**<font color = 'purple'> [e.g., height difference between 8-year old girls and 18-year old girls]</font>: Requires a small _N_ (e.g., _N is approximately 26_).

##💡🔆 Extra information:
**Cohen’s _d_ (The Comparison Ruler)**

- Used for: Comparing the means of two groups (e.g., _t_-tests).
- Metric: Standardized Mean Difference.
- Example: The difference between your students' scores in Session 1 vs. Session 2


**Cohen’s _f^2_[f-squared] (The Variance Ruler)**
- Used for: Multiple Regression and Mixed-Effects Models (LMER).
- Metric: Proportion of Variance.
- Example: How much of the change in pronunciation is caused by the Interaction between Session and Repetition.

**Why did he invent both?**
- Cohen realized that one "ruler" doesn't fit every experiment:_d_ is like a measuring tape used to find the distance between two points (Group A and Group B).
- _f^2_ is like a pie chart used to see how much of a "result pie" is eaten by one specific ingredient (Variable X).

**Other "Cohen" contributions while _d_ and _f^2_ are his most famous, he also gave us:
- Cohen’s _h_: For comparing proportions (e.g., the % of students who passed vs. failed).
- Cohen’s k (Kappa): For Inter-rater Reliability.
- Power Analysis: He literally wrote the book on how to calculate _N_ before you start an experiment.
